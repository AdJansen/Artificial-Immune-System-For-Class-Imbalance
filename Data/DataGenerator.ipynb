```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.datasets import make_classification
import matplotlib.colors
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python

# Generate Clean data with all features informative and a larger class seperation
X,y = make_classification(n_samples=300, n_features=5, n_informative=5,n_redundant=0, n_repeated=0, n_classes=2, n_clusters_per_class=1,class_sep=100,flip_y=0,weights=[0.7,0.3])

#Display the graph of the generated data
f, ax1 = plt.subplots(nrows=1, ncols=1,figsize=(20,8))
sns.scatterplot(X[:,0],X[:,1],hue=y,ax=ax1)

#Set the title
ax1.set_title("NosielessInformativeEasy")

#Show the plot
plt.show()

#Send the data to a csv file
X_data = np.array(X)
y_data = np.array(y)
new_data = np.append(X_data ,y_data.reshape(-1,1),axis=1)
df = pd.DataFrame(new_data)
df.to_csv('GeneratedSyntheticData-NosielessInformativeEasy.csv')
```
